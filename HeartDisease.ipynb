```python
q1=df.trestbps.mean()-0.675*df.trestbps.std()
q3=df.trestbps.mean()+0.675*df.trestbps.std()
iqr=q3-q1
max=q3+1.5*iqr
min=q1-1.5*iqr

from sklearn.preprocessing import StandardScaler,RobustScaler,MinMaxScaler
import scipy.stats as stat
import pylab
import numpy as np
scalers={'StandardScaler':StandardScaler,'RobustScaler':RobustScaler,'MinMaxScaler':MinMaxScaler}
def plot_data(df,feature,key):
  plt.figure(figsize=(15,6))
  plt.subplot(1,3,1)
  df[feature].hist()
  plt.subplot(1,3,2)
  stat.probplot(df[feature],dist="norm",plot=pylab)
  plt.subplot(1,3,3)
  sns.boxplot(df[feature])
  plt.title(feature+'__'+key)
  plt.show()

def transformations(df,feature):

  for key,scaler in scalers.items():
    scaler_method=scaler()
    df_scaled=pd.DataFrame(scaler_method.fit_transform(df),columns=df.columns)
    plot_data(df_scaled,feature,key)
  
  # Log Transformation
  df_scaled[feature+'_log']=np.log1p(df[feature])
  plot_data(df_scaled,feature+'_log',"log")

  # Reciprocal Transformation
  try:
    df_scaled[feature+'_Reciprocal']=np.reciprocal(df[feature])
    plot_data(df_scaled,feature+'_Reciprocal','Reciprocal')
  except:
    pass
  # Square Root Transformation
  df_scaled[feature+'sqrt']=(df[feature])**(1/2)
  plot_data(df_scaled,feature+'sqrt','sqrt')

  # Exponential Transformation
  df_scaled[feature+'_exponential']=(df[feature])**(1/1.2)
  plot_data(df_scaled,feature+'_exponential','exponential')

  df_scaled[feature+'_boxcox'],parameters=stat.boxcox(df[feature])
  plot_data(df_scaled,feature+'_boxcox','boxcox')

  del df_scaled
```

```python
a = pd.get_dummies(df['cp'], prefix = "cp")
# b = pd.get_dummies(df['thal'], prefix = "thal")
c = pd.get_dummies(df['slope'], prefix = "slope")
frames = [df, a,c]
df = pd.concat(frames, axis = 1)
df = df.drop(columns = ['cp', 'slope'])
df.head()
```

```python
df.target.value_counts()
```

```python
X=df.drop('target',axis=1)
y=df['target']
X=pd.DataFrame(StandardScaler().fit_transform(X),columns=X.columns)
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(max_depth=5,n_estimators = 1000, random_state = 1)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print('accuracy_score ==> ','\n',accuracy_score(y_test,y_pred))
print('confusion_matrix ==> ','\n',confusion_matrix(y_test,y_pred))
print('classification_report ==> ','\n',classification_report(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 2)  # n_neighbors means k
knn.fit(X_train, y_train)
prediction = knn.predict(X_test)

scoreList = []
for i in range(1,20):
    knn2 = KNeighborsClassifier(n_neighbors = i)  # n_neighbors means k
    knn2.fit(X_train, y_train)
    scoreList.append(knn2.score(X_test, y_test))
    
plt.plot(range(1,20), scoreList)
plt.xticks(np.arange(1,20,1))
plt.xlabel("K value")
plt.ylabel("Score")
plt.show()

acc = np.max(scoreList)*100
print(scoreList)
# accuracies['KNN'] = acc
print("Maximum KNN Score is {:.2f}%".format(acc))
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print('accuracy_score ==> ','\n',accuracy_score(y_test,y_pred))
print('confusion_matrix ==> ','\n',confusion_matrix(y_test,y_pred))
print('classification_report ==> ','\n',classification_report(y_test,y_pred))
```
